## Computing Common Neighbors Score

We define the "common neighbours" link score:

```python
def common_neighbors(G,i,j):
    # G - the graph
    # i,j - the link
    return len(set(G.neighbors(i)) and set(G.neighbors(j)))
```

```python
common_neighbors(G,'31','17')
```

```python
def jaccard_coef(G, i, j):
    return len((set(G.neighbors(i)) and set(G.neighbors(j))))/len((set(G.neighbors(i)) or set(G.neighbors(j))))
```

```python
jaccard_coef(G,'31','17')
```

```python
def inverse_distance(G, i, j):
    if i == j:
        return 0
    return nx.resistance_distance(G, i, j)
```

```python
inverse_distance(G,'31','17')
```

```python
def preferential_attachment(G, i, j):
    ki = G.degree(i)
    kj = G.degree(j)
    return ki*kj
```

## Link Prediction

We define a function which returns a sorted list of of scores from a node to other possible nodes in the graph:

```python
def link_list(G,i,score_func):
    # G - the graph
    # i - the node
    # score_func - the link scoring function, having the signature score_func(G,i,j)
    links = []
    for j in G.nodes():
        if not G.has_edge(i,j):
            e = (i,j)
            sc = score_func(G,i,j)
            links.append([e,sc])
    links.sort(key = lambda x: x[1], reverse = True) 
    return links
```

```python
link_list(G,'31',common_neighbors)
```

```python
link_list(G,'31',jaccard_coef)
```

```python
link_list(G,'31', inverse_distance)
```

```python
link_list(G,'31', preferential_attachment)
```
